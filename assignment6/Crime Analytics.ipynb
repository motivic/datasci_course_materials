# Theft of Motor Vehicle Contents or Parts (Car prowl) is the Most Common Type of Criminal Activity in the Summer in Both Seattle and San Francisco.

Moreover, the majority of car prowls take place between 3PM-1AM.

## Top Ten Criminal Activities

To begin, we look at the top ten criminal activities (using Descript/Summarized Offense Description) in each city. From the bar charts below, we see that the most common crime in Seattle is Theft-Carprowl, and that in San Francisco is Grand Theft from Locked Auto. Also, we color the bars based on Category/Offense Type, from which we see the highest occurring crime in Seattle is in the Car Prowl category, and that in San Francisco is in Larceny/Theft, which is the equivalent of Car Prowl as infered from the descriptions of the actual criminal activities.

```python
# References:
# http://www.machinalis.com/blog/embedding-interactive-maps-into-an-ipython-nb/
# http://protips.maxmasnick.com/hide-code-when-sharing-ipython-notebooks

%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib as mpl
import pandas as pd
import numpy as np

from collections import Counter, OrderedDict
from datetime import datetime

mpl.style.use('ggplot')
mpl.pylab.rcParams['figure.figsize'] = 16, 10
```

```python
# Read in the columns that we will use in our data analysis
sf_fields = ['Category', 'Descript', 'Date', 'Time', 'Location']
sea_fields = ['Summarized Offense Description', 'Offense Type', 'Occurred Date or Date Range Start', 'Location']

df_sea = pd.read_csv('seattle_incidents_summer_2014.csv', usecols=sea_fields)
df_sf = pd.read_csv('sanfrancisco_incidents_summer_2014.csv', usecols=sf_fields)

# relabel Seattle data to match the labels used for San Francisco
df_sea.columns = ['Descript', 'Category', 'Date', 'Location']

# replace "Date" and "Time" with into a new column "Timestamp" of type <datetime>
format_sea_time = lambda x: datetime.strptime(x, "%m/%d/%Y %I:%M:%S %p")
df_sea['Timestamp'] = pd.Series(df_sea['Date'].map(format_sea_time))
df_sea = df_sea.drop('Date', 1)

format_sf_date = lambda x: datetime.strptime(x, "%m/%d/%Y %H:%M")
format_sf_time = lambda x: datetime.strptime(x, "%H:%M")
df_sf['Timestamp'] = pd.Series((df_sf['Date']+' '+df_sf['Time']).map(format_sf_date))
df_sf = df_sf.drop('Date', 1)
df_sf = df_sf.drop('Time', 1)
```

```python
sea_crime_count = OrderedDict(sorted(Counter(df_sea['Descript']).items(), key=lambda t: t[1], reverse=True))
sf_crime_count = OrderedDict(sorted(Counter(df_sf['Descript']).items(), key=lambda t: t[1], reverse=True))

sea_cat_desc_dict = dict(zip(df_sea['Descript'], df_sea['Category']))
sf_cat_desc_dict = dict(zip(df_sf['Descript'], df_sf['Category']))

# Setup colors for different categories
cmap = plt.get_cmap('rainbow')

# Uniqify list
def uniquify(seq): 
    # order preserving
    noDupes = []
    [noDupes.append(i) for i in seq if not noDupes.count(i)]
    return noDupes

sea_top_crimes = uniquify([sea_cat_desc_dict[v] for v in sea_crime_count.keys()[:10]])
sf_top_crimes = uniquify([sf_cat_desc_dict[v] for v in sf_crime_count.keys()[:10]])
sea_colors = ['maroon', 'red', 'darkorange', 'tan', 'darkblue', 'rosybrown', 'g', 'salmon', 'orangered']
sf_colors = ['maroon', 'red', 'burlywood', 'skyblue', 'darkblue', 'yellowgreen']
sea_d2c = OrderedDict(zip(sea_top_crimes, sea_colors))
sf_d2c = OrderedDict(zip(sf_top_crimes, sf_colors))

def desc2color(cat_desc_dict, d2c, desc):
    return d2c[cat_desc_dict[desc]]

sea_top_colors = [desc2color(sea_cat_desc_dict, sea_d2c, x) for x in sea_crime_count.keys()[:10]]
sf_top_colors = [desc2color(sf_cat_desc_dict, sf_d2c, x) for x in sf_crime_count.keys()[:10]]
```

```python
# Plot the top ten crimes in each city
f, (ax1, ax2) = plt.subplots(1, 2)

# Seattle
rects1 = ax1.bar(range(0,80,8), sea_crime_count.values()[:10], color=sea_top_colors, width=4)
ax1.set_title('Seattle')
ax1.get_yaxis().set_visible(False)
ax1.set_xticks(range(2,82,8))
ax1.set_xticklabels(sea_crime_count.keys(), rotation='vertical', fontsize=13)
legend1 = [mpatches.Patch(color=c, label=crime) for (crime, c) in sea_d2c.items()] 
ax1.legend(handles=legend1)

# San Francisco
rects2 = ax2.bar(range(0,80,8), sf_crime_count.values()[:10], color=sf_top_colors, width=4)
ax2.set_title('San Francisco')
ax2.get_yaxis().set_visible(False)
ax2.set_xticks(range(2,82,8))
ax2.set_xticklabels(sf_crime_count.keys(), rotation='vertical', fontsize=13)
legend2 = [mpatches.Patch(color=c, label=crime) for (crime, c) in sf_d2c.items()] 
ax2.legend(handles=legend2)

def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height, '%d' % int(height), ha='center', va='bottom', fontsize=13)

autolabel(rects1, ax1)
autolabel(rects2, ax2)
plt.suptitle("Top Ten Crimes", fontsize=24)
plt.draw()
```

## Top Criminal Activities by Month

We also looked into the number of criminal activities in each of the top categories above in each of the months June, July, and August in each city. Car prowl is again the dominant type of criminal activity across all three months and in both cities.

We also observed an interesting trend that the number of crime activities has decreased monthly in Seattle but has increased every month in San Francisco.

```python
# Plot the top crime occurrences based on month

sea_topcrimes = df_sea[df_sea.Descript.isin(sea_crime_count.keys()[:10])].copy(deep=True)
sf_topcrimes = df_sf[df_sf.Descript.isin(sf_crime_count.keys()[:10])].copy(deep=True)

# Seattle
fig, ax = plt.subplots()
ax1 = ax.twinx()
format_sea_month = lambda x: datetime.strftime(x, "%m")
sea_topcrimes['Month'] = pd.Series(sea_topcrimes['Timestamp'].map(format_sea_month))

# Plot monthly total crime
sea_monthly_total = sea_topcrimes[['Month', 'Timestamp']].groupby(['Month']).agg(['count']).unstack()
sea_monthly_total.plot(ax=ax1, kind='bar', color='pink', alpha=0.3)
#ax1.legend(loc=0, prop={'size':10})

# Plot monthly bar chart by crime
sea_count = sea_topcrimes[['Month', 'Category', 'Timestamp']].groupby(['Month', 'Category']).agg(['count']).unstack()
sea_count.columns = [cat for _,_,cat in sea_count.columns]
sea_count = sea_count[sea_top_crimes]
sea_count.plot(ax=ax, kind='bar', color=sea_colors);
ax.set_title('Top Criminal Activities by Month in Seattle')
ax.set_xticklabels(['June', 'July', 'August'], rotation='horizontal')
legend_sea = [mpatches.Patch(color=c, label=crime) for (crime, c) in sea_d2c.items()]
legend_sea.append(mpatches.Patch(color='pink', label='TOTAL'))
ax.legend(handles=legend_sea, loc=0, ncol=2, prop={'size':10})
plt.show()

# San Francisco
fig, ax = plt.subplots()
ax1 = ax.twinx()
format_sf_month = lambda x: datetime.strftime(x, "%m")
sf_topcrimes['Month'] = pd.Series(sf_topcrimes['Timestamp'].map(format_sf_month))

# Plot monthly total crime
sf_monthly_total = sf_topcrimes[['Month', 'Timestamp']].groupby(['Month']).agg(['count']).unstack()
sf_monthly_total.plot(ax=ax1, kind='bar', color='pink', alpha=0.3)

# Plot monthly bar chart by crime
sf_count = sf_topcrimes[['Month', 'Category', 'Timestamp']].groupby(['Month', 'Category']).agg(['count']).unstack()
sf_count.columns = [cat for _,_,cat in sf_count.columns]
sf_count = sf_count[sf_top_crimes]
sf_count.plot(ax=ax, kind='bar', color=sf_colors);
ax.set_title('Top Criminal Activities by Month in San Francisco')
ax.set_xticklabels(['June', 'July', 'August'], rotation='horizontal')
legend_sf = [mpatches.Patch(color=c, label=crime) for (crime, c) in sf_d2c.items()]
legend_sf.append(mpatches.Patch(color='pink', label='TOTAL'))
ax.legend(handles=legend_sf, loc=0, ncol=2, prop={'size':10})

plt.show()
```

## Daily Criminal Activity Trends

Finally, we look at how the number of criminal activities change over the day. In both cities, the hours between 3AM-5:30AM show the lowest rate of criminal activities, while the hours between 8PM-11PM show the highest rate of crimianl activities. Also, there is spike in criminal activity around noon time in both cities. The trends also holds on a categorical basis with the except of the categories Burglary, Other Property, and Shoplifting in Seattle, which are more frequent earlier in the day (between 7AM-12PM).

```python
# Plot the top crime occurrences during the day

sea_topcrimes = df_sea[df_sea.Descript.isin(sea_crime_count.keys()[:10])].copy(deep=True)
sf_topcrimes = df_sf[df_sf.Descript.isin(sf_crime_count.keys()[:10])].copy(deep=True)

# Seattle
format_sea_hour = lambda x: datetime.strftime(x, "%H")
sea_topcrimes['Hour'] = pd.Series(sea_topcrimes['Timestamp'].map(format_sea_hour))
sea_count = sea_topcrimes[['Hour', 'Category', 'Timestamp']].groupby(['Hour', 'Category']).agg(['count']).unstack()
sea_count.columns = [cat for _,_,cat in sea_count.columns]
c = OrderedDict(sorted(sea_d2c.items())).values()
ax = sea_count.plot(kind='area', color=c);
ax.set_title('Daily Trend of Top Criminal Activities in Seattle')

plt.show()

# San Francisco
format_sf_hour = lambda x: datetime.strftime(x, "%H")
sf_topcrimes['Hour'] = pd.Series(sf_topcrimes['Timestamp'].map(format_sf_hour))
sf_count = sf_topcrimes[['Hour', 'Category', 'Timestamp']].groupby(['Hour', 'Category']).agg(['count']).unstack()
sf_count.columns = [cat for _,_,cat in sf_count.columns]
c = OrderedDict(sorted(sf_d2c.items())).values()
ax = sf_count.plot(kind='area', color=c);
ax.set_title('Daily Trend of Top Criminal Activities in San Francisco')

plt.show()
```
